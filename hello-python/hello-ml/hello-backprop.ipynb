```python
import numpy as np
```

```python
# Just a really simple example of using gradient descent to find a line that separates two classes.
# The examples are simply points separated by a diagonal line through the origin, with labels 1 and -1.
# Run the cell and see that w converges to (1,1) and b close to zero.

X = np.array([[1, 0],
              [0, 1],
              [-1, 0],
              [0, -1]])

y = np.array([1,
              1,
             -1,
             -1])

w = [1, 0]
b = 0.5
eta = 0.1
epochs = 250

grads = []
thetas = []
for j in range(epochs):
    wgrad = []
    bgrad = []
    z = []
    for i in range(len(X)):
        xi = X[i]
        yi = y[i]
        zi = np.dot(xi, w) + b

        dLdz = zi - yi
        dLdw = dLdz * xi
        dLdb = dLdz * 1
        wgrad.append(dLdw)
        bgrad.append(dLdb)
        z.append(zi)
        # print("dLdw:", dLdw, "dLdb:", dLdb)

    L = np.mean((y - z)**2) / 2
    wgrad = np.mean(wgrad, axis=0)
    bgrad = np.mean(bgrad)
    w = w - eta * wgrad
    b = b - eta * bgrad
    grad = np.concatenate((wgrad, [bgrad]))
    theta = np.concatenate((w, [b]))
    grads.append(grad)
    thetas.append(theta)
    print(L, w, b, wgrad, bgrad)
# print(13, "gradnorm:", )
```
